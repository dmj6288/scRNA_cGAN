```python

import scanpy as sc
import matplotlib.pyplot as plt
import scrublet as scr
import numpy as np
import h5py
from scipy.sparse import csc_matrix
import decoupler as dc
import pandas as pd
import anndata as ad
import scvi
```

```python

adata = sc.read_h5ad("../ZafrensData/merged_pseudobulk_adata.h5ad")
```

```python
adata
```

```python

# Basic preprocessing (if needed)
sc.pp.normalize_total(adata, target_sum=1e4)  # Normalize gene counts
sc.pp.log1p(adata)                            # Log-transform the data
sc.pp.highly_variable_genes(adata, n_top_genes=5000, subset=True)
```

```python

# Register AnnData with scvi-tools
scvi.model.SCVI.setup_anndata(adata)

# Define and initialize the VAE model
vae_model = scvi.model.SCVI(adata, n_latent=2000)  # Change n_latent if needed
```

```python
# Train the model
vae_model.train(max_epochs=200, early_stopping=True, logger=False)

# Check the training progress (optional)
#vae_model.history
```

```python

# Extract the latent space embeddings for each cell
latent_embeddings = vae_model.get_latent_representation()

# Convert to a DataFrame if needed
latent_df = pd.DataFrame(latent_embeddings, index=adata.obs_names)
latent_df.to_csv("latent_embeddings.csv", index=True)
```

```python

# Add latent space to AnnData
adata.obsm['X_latent'] = latent_embeddings

# Run UMAP on the latent embeddings
sc.pp.neighbors(adata, use_rep='X_latent')
sc.tl.umap(adata)

# Plot UMAP visualization
sc.pl.umap(adata)  # Replace 'cell_type' with a relevant metadata column
```
